### Unstructured Loader

i got a error here kernel restarting automatically thats Why used another approach

```python
from langchain_community.document_loaders import UnstructuredURLLoader

loader = UnstructuredURLLoader(
    urls=["https://www.cnbc.com/2024/12/21/how-the-federal-reserves-rate-policy-affects-mortgages.html"]
)
docs = loader.load()
len(docs)
docs[0].metadata
print(docs[0].page_content)
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.cnbc.com/2024/12/21/how-the-federal-reserves-rate-policy-affects-mortgages.html")
docs = loader.load()
print(len(docs))
```

```python
docs[0].metadata
```

```python
print(docs[0].page_content)
```
